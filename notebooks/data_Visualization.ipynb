```python
import pandas as pd 
import numpy as np 
from plotly.subplots import make_subplots
import plotly.express as px
import plotly.graph_objects as go
from math import ceil
```

```python
class VizData:
    def __init__(self, data: pd.DataFrame):
        self.data = data.copy()

    def plotfeatures(self, freq_min=60, title="Features"):
        features = self.data.columns.tolist()
        df_resampled = self.data.groupby(pd.Grouper(freq=f"{freq_min}T")).mean()
        ncols = 2
        nrows = ceil(len(features)/ncols)
        fig = make_subplots(rows=nrows, cols=ncols, subplot_titles=features)
        for i, f in enumerate(features):
            row = i // ncols + 1
            col = i % ncols + 1
            fig.add_trace(go.Scatter(x=df_resampled.index, y=df_resampled[f], mode='lines', name=f), row=row, col=col)
        fig.update_layout(height=600, width=1000, title_text=title, showlegend=False)
        fig.show()    
```
